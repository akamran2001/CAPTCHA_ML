```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# Load the image
image = Image.open("captchas/recaptcha/rc-image-tile-33/captcha_0.png")

# Convert the image to a numpy array
image_array = np.array(image)

# Split the image array into 9 100x100 sections
sections = []
for i in range(3):
    for j in range(3):
        section = image_array[i*100:(i+1)*100, j*100:(j+1)*100]
        sections.append(section)

# Convert the sections list to a numpy array
sections_array = np.array(sections)
sections_array.shape
```

```python
# Display each image in the array
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(6,6))
for i, ax in enumerate(axes.flat):
    ax.imshow(sections_array[i], cmap='gray')
    ax.set_axis_off()
plt.tight_layout()
plt.show()
```
