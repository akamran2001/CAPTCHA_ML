```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import tensorflow as tf
from tensorflow import keras as ks
```

```python
class_names = {}

# Define the image size and batch size
img_size = (200, 50)

# Load the image data
image_paths = []
labels = []

dataset_dir = 'captchas'
for i, filename in enumerate(os.listdir(dataset_dir)):
    if filename.endswith('.png'):
        image_paths.append(os.path.join(dataset_dir, filename))
        label = filename.split('.')[0].lower()
        labels.append(i)
        class_names[i] = label

images = []
for path in image_paths:
    image = Image.open(path)
    image = image.resize(img_size)
    image = np.array(image)
    images.append(image)

images = np.array(images)
labels = np.array(labels)

# Split the data into training and validation sets
num_train = int(0.85 * len(images))

train_images = images[:num_train]
train_labels = labels[:num_train]
val_images = images[num_train:]
val_labels = labels[num_train:]

num_classes = labels.max() + 1
```

```python
# Show a sample of images and labels
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[i])
plt.show()
```

```python
print('Training images shape: ', train_images.shape)
```

```python
# Define the model architecture
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Reshape, BatchNormalization

model = ks.Sequential([
    Flatten(input_shape=train_images.shape[1:4]),
    BatchNormalization(),
    Dense(512, activation='relu'),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])

# Train the model
model.fit(
    train_images,
    train_labels,
    batch_size=100,
    epochs=20,
    validation_split=0.15,
    validation_data=(val_images, val_labels))
```

```python
# Show a sample of images, their predicted labels, and true labels
plt.figure(figsize=(10, 10))

num_samples = 16
sqr = int(np.sqrt(num_samples))
indices = np.random.choice(np.arange(len(val_images)), size=num_samples, replace=False)

sample_images = val_images[indices]
sample_labels = val_labels[indices]

for i in range(num_samples):
    plt.subplot(sqr, sqr, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(sample_images[i], cmap=plt.cm.binary)
    pred_label = np.argmax(model.predict(np.expand_dims(sample_images[i], axis=0)))
    true_label = sample_labels[i]
    plt.xlabel('pred: {},\ntrue: {}'.format(class_names[pred_label], class_names[true_label]))
    
    # Draw a green border around the image if the prediction is correct and a red border if it is incorrect
    if pred_label == true_label:
        plt.gca().add_patch(plt.Rectangle((0, 0), img_size[0], img_size[1], fill=False, edgecolor='green', linewidth=3))
    else:
        plt.gca().add_patch(plt.Rectangle((0, 0), img_size[0], img_size[1], fill=False, edgecolor='red', linewidth=3))

# Show the model's evauluation on the validation set as a label on the plot
score = model.evaluate(val_images, val_labels, verbose=0)
plt.suptitle('Model accuracy: {:.2f}%'.format(score[1] * 100))
plt.show()
```
